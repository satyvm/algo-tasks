```python
import backtesting
from backtesting import Backtest, Strategy
from backtesting.lib import crossover

from backtesting.test import SMA, GOOG
```

```python
print(GOOG)
```

```python
class sma_cross(Strategy):
    def init(self):
        price = self.data.Close
        self.ma1 = self.I(SMA, price, 10)
        self.ma2 = self.I(SMA, price, 20)

    def next(self):
        if crossover(self.ma1, self.ma2):
            self.buy()
        elif crossover(self.ma2, self.ma1):
            self.sell()


bt = Backtest(GOOG, sma_cross, commission=.002, exclusive_orders=True)
stats = bt.run()
print(stats)
bt.plot()
```

```python
from yfinance import download
```

```python
symbols = ["GOOG", "AAPL", "MSFT"]
start_date = "2020-01-01"
end_date = "2024-09-16"

data = {}
for symbol in symbols:
    data[symbol] = download(symbol, start=start_date, end=end_date)
```

```python
print(data["GOOG"])
```

```python
def sma_backtest(symbol, commission=0.002, exclusive_orders=True):
    price = data[symbol]
    bt = Backtest(price, sma_cross, commission=commission, exclusive_orders=exclusive_orders)
    stats = bt.run()
    bt.plot()
    return stats
```

```python
for symbol in symbols:
    stats = sma_backtest(symbol, commission=0.002, exclusive_orders=True)
    print(f"Backtest results for {symbol}:\n", stats)
```
