```python
def get_account_balance():
    header={
        "accept": "application/json",
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    acc = requests.get("https://paper-api.alpaca.markets/v2/account", headers=header)
    return acc.json()['buying_power']
print(get_account_balance())
```

```python
def get_open_positions():
    url = "https://paper-api.alpaca.markets/v2/positions"
    
    header = {
        "accept": "application/json",
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    
    response = requests.get(url, headers=header)
    
    print(response.text)
get_open_positions()
```

```python
def get_order_history():
    url = "https://paper-api.alpaca.markets/v2/orders?status=all"
    
    header = {
        "accept": "application/json",
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    response = requests.get(url, headers=header)
    data = response.json()
    for j in range(len(data)):
        print(data[j].get('id'))
get_order_history()
```

```python
now = datetime.now(ZoneInfo('Asia/Kolkata'))
print(now)
print(now - timedelta(days = 1))
```

```python
def get_market_data(symbol, start= (now - timedelta(days = 1)).isoformat(), end=now.isoformat(), timeframe="1Min"):
    url = "https://data.alpaca.markets/v1beta3/crypto/us/bars"
    params = {
        "symbols": symbol,
        "timeframe": timeframe,
        "start": start,
        "end": end,
        "sort": "asc"
    }
    
    # Define the headers
    header = {"accept": "application/json"}
    response = requests.get(url, headers=header, params=params)
    data = response.json()
    print(data)
    return data
market_data = get_market_data('BTC/USD')
for i in range(len(market_data['bars']['BTC/USD'])):
    print(market_data['bars']['BTC/USD'][i]['c'], "-", market_data['bars']['BTC/USD'][i]['t'])
```

```python
def get_realtime_quote(symbol):
    url = "https://data.alpaca.markets/v1beta3/crypto/us/latest/quotes"

    header = {"accept": "application/json"}
    
    params = {
        "symbols": symbol
    }
    
    response = requests.get(url, headers=header, params=params)
    
    print(response.json()['quotes'])
get_realtime_quote('BTC/USD')
```

```python
from alpaca.trading.client import TradingClient
from alpaca.data.timeframe import TimeFrame, TimeFrameUnit
from alpaca.data.historical.crypto import CryptoHistoricalDataClient

from alpaca.data.requests import (
    CryptoBarsRequest,
    CryptoLatestQuoteRequest
    )
from alpaca.trading.requests import (
    GetOrdersRequest,
)
from alpaca.trading.enums import (
    QueryOrderStatus
)
```

```python
paper = True
now = datetime.now(ZoneInfo('Asia/Kolkata'))
```

```python
trading_client = TradingClient(API_KEY, SECRET_KEY, paper=paper)
crypto_historical_data_client = CryptoHistoricalDataClient()
```

```python
def get_account_balance_2():
    acc = trading_client.get_account()
    if acc.trading_blocked:
        print('Account is currently restricted from trading.')

    print('${} is available as buying power.'.format(acc.buying_power))
get_account_balance_2()
```

```python
def get_all_positions_2(client):
    portfolio = client.get_all_positions()

    for position in portfolio:
        print("{} shares of {}".format(position.qty, position.symbol))
        
    return portfolio
print(get_all_positions_2(trading_client))
```

```python
def get_order_history_2(client):
    request_params = GetOrdersRequest(
                    status=QueryOrderStatus.ALL
                 )
    order_history = client.get_orders(filter=request_params)
    print("Order History: ", order_history)
    
    return order_history
print(get_order_history_2(trading_client))
```

```python
def get_market_data_2(symbol, start= now - timedelta(days = 1), end=now, limit=None):
    req = CryptoBarsRequest(
        symbol_or_symbols = [symbol],
        timeframe=TimeFrame(amount = 1, unit = TimeFrameUnit.Hour), # specify timeframe
        start = start,
        end_date= end,
        limit = limit,
    )
    return crypto_historical_data_client.get_crypto_bars(req).df
print(get_market_data_2('BTC/USD'))
```

```python
def get_realtime_quote_2(symbol):
    req = CryptoLatestQuoteRequest(
        symbol_or_symbols = [symbol],
    )
    return crypto_historical_data_client.get_crypto_latest_quote(req)
print(get_realtime_quote_2('BTC/USD'))
```

## Again with REST APIs

```python
def place_order(symbol, quantity, order_type, side, limit_price=None):
    url = "https://paper-api.alpaca.markets/v2/orders"
    payload = {
        "side": side,
        "type": order_type,
        "time_in_force": "gtc",
        "symbol": symbol,
        "qty": quantity,
        "limit_price": limit_price
    }
    header = {
        "accept": "application/json",
        "content-type": "application/json",
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    response = requests.post(url, json=payload, headers=header)
    print(response.text)
    return response.json()['id']
new_id = place_order("BTC/USD", 0.1, "market", "buy")
```

```python
def modify_order(order_id, new_quantity):
    url = f"https://paper-api.alpaca.markets/v2/orders/{order_id}"

    payload = {
        "qty": new_quantity,
    }
    header = {
        "accept": "application/json",
        "content-type": "application/json",
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    
    response = requests.patch(url, json=payload, headers=header)
    print(response.text)
modify_order("74518f56-08ea-4d31-81fe-4900ea83f245", 0.03)
```

```python
def cancel_order(order_id):
    url = f"https://paper-api.alpaca.markets/v2/orders/{order_id}"

    header = {
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    
    response = requests.delete(url, headers=header)
    print(response.status_code)

new_order_id = place_order("ETH/USD", 0.01, "limit", "buy", 2100)
cancel_order(new_order_id)
```

```python
def place_limit_order(symbol, quantity, side, limit_price):
    url = "https://paper-api.alpaca.markets/v2/orders"
    payload = {
        "side": side,
        "type": "limit",
        "time_in_force": "gtc",
        "symbol": symbol,
        "qty": quantity,
        "limit_price": limit_price
    }
    header = {
        "accept": "application/json",
        "content-type": "application/json",
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    response = requests.post(url, json=payload, headers=header)
    print(response.text)
    return response.json()['id']
new_order_id = place_limit_order("BTC/USD", 0.1, "buy", "59561")
```

```python
def place_stop_order(symbol, quantity, side, limit_price, stop_price):
    url = "https://paper-api.alpaca.markets/v2/orders"
    payload = {
        "side": side,
        "type": "stop_limit",
        "time_in_force": "gtc",
        "symbol": symbol,
        "qty": quantity,
        "limit_price": limit_price,
        "stop_price": stop_price
    }
    header = {
        "accept": "application/json",
        "content-type": "application/json",
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    response = requests.post(url, json=payload, headers=header)
    print(response.text)
    return response.json()['id']
new_stop_order_id = place_stop_order("BTC/USD", 0.1, "buy", "59540", "59538")
```

```python
def get_order_status(order_id):
    url = f"https://paper-api.alpaca.markets/v2/orders/{order_id}"
    
    header = {
        "accept": "application/json",
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    response = requests.get(url, headers=header)
    
    if response.status_code == 200:
        order_data = response.json()
        
        if order_data.get('filled_at') is not None:
            print("Order fulfilled at ", order_data['filled_at'])
        elif order_data.get('expired_at') is not None:
            print("Order expired at ", order_data['expired_at'])
        elif order_data.get('canceled_at') is not None:
            print("Order canceled at ", order_data['canceled_at'])
        elif order_data.get('failed_at') is not None:
            print("Order failed at ", order_data['failed_at'])
        elif order_data.get('replaced_at') is not None:
            print("Order replaced at ", order_data['replaced_at'], " by ", order_data['replaced_by'])
        else:
            print("Order is still pending")
    else:
        print("Failed to fetch order status")
get_order_status("74518f56-08ea-4d31-81fe-4900ea83f245")
```

# Logging and Error Handling

```python
import logging
```

```python
logging.basicConfig(
    filename='order_status.log', 
    level=logging.INFO, 
    format='%(asctime)s - %(levelname)s - %(message)s'
)
```

```python
def log_api_call(endpoint, status_code, response_time):
    logging.info(f"API Call to {endpoint} returned status code {status_code} in {response_time:.2f} seconds")

def handle_api_error(response):
    if response.status_code != 200:
        logging.error(f"API Error: Status Code {response.status_code} - {response.text}")
```

```python
def get_order_status(order_id):
    url = f"https://paper-api.alpaca.markets/v2/orders/{order_id}"
    
    header = {
        "accept": "application/json",
        "APCA-API-KEY-ID": API_KEY,
        "APCA-API-SECRET-KEY": SECRET_KEY
    }
    start_time = datetime.now()
    response = requests.get(url, headers=header)
    response_time = (datetime.now() - start_time).total_seconds()
    
    log_api_call(url, response.status_code, response_time)
    handle_api_error(response)
    
    if response.status_code == 200:
        order_data = response.json()
        if order_data.get('filled_at') is not None:
            print("Order fulfilled at ", order_data['filled_at'])
        elif order_data.get('expired_at') is not None:
            print("Order expired at ", order_data['expired_at'])
        elif order_data.get('canceled_at') is not None:
            print("Order canceled at ", order_data['canceled_at'])
        elif order_data.get('failed_at') is not None:
            print("Order failed at ", order_data['failed_at'])
        elif order_data.get('replaced_at') is not None:
            print("Order replaced at ", order_data['replaced_at'], " by ", order_data['replaced_by'])
        else:
            print("Order is still pending")
    else:
        print("Failed to fetch order status")
get_order_status("74518f56-08ea-4d31-81fe-4900ea83f245")
```
